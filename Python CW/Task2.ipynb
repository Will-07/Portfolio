```python
import numpy as np
import matplotlib.pyplot as plt
#initialize the grid
x = np.arange(-10,10,1)
y = np.arange(-10,10,1)

X,Y = np.meshgrid(x,y)
#ferromagnetism
xd1 = 0
yd1 = 1
plt.quiver(X,Y,xd1,yd1,color='purple')
plt.title('ferromagnetism')
plt.xlabel('X coord')
plt.ylabel('Y coord')
plt.show()

#antiferromagnetism
xd2 = 0
yd2 = (-1)**(np.abs(X))
plt.quiver(X,Y,xd2,yd2,color='green')
plt.title('antiferromagnetism')
plt.xlabel('X coord')
plt.ylabel('Y coord')
plt.show()

#new grid in radians to show spin magnetism better
x3 = np.arange(-2*np.pi,2*np.pi,np.pi/4)
y3 = np.arange(-2*np.pi,2*np.pi,np.pi/4)

X3,Y3 = np.meshgrid(x3,y3)

#spin magnetism
xd3 = np.cos((X3+Y3))
yd3 = np.sin((X3+Y3))
plt.quiver(X3,Y3,xd3,yd3,color='red')
plt.title('spin magnetism')
plt.xlabel('X coord')
plt.ylabel('Y coord')
plt.show()
```
