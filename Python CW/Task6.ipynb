```python
import matplotlib.pyplot as plt
import numpy as np

c = 3e8
h = 6.63e-34
k = 1.38e-23
B = []
maxB = []
maxlamda = []
Wiens_law_temp = []
Actual_temp = []
for T  in range(5000, 50000, 5000):
    lamda = np.arange(0.5e-9, 0.1e-5, 0.1e-9)
    B = ((2*h*c**2)/(lamda**5))/(np.exp((h*c)/(lamda*k*T))-1)
    maxB = np.argmax(B)
    maxlamda.append(lamda[maxB])
    plt.plot(lamda,B)
    plt.xlabel('Wavelength (m)')
    plt.ylabel('Spectral radiance (Wm^-2nm^-1)')
print(maxlamda)
plt.show()
#plots B as lamda(written lamda in code as lambda exists as some function in python) is varied, to find a lamda where B is at a
#maximum.
Actual_temp = np.arange(5000, 50000, 5000)
for i in range (len(maxlamda)):
    Wiens_law_temp.append(0.002898/maxlamda[i])
print('Actual Temp =',Actual_temp)
print('Weins law Temp =',Wiens_law_temp)
#shows the temperature input, and the temperature calculated from Wiens law.
```
