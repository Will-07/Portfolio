```python
import matplotlib.pyplot as plt
import numpy as np
import random
#function for making a random walk
def random_walk(Nsteps):
    x = []
    y = []
    x.append(0)
    y.append(0)
    xc = 0
    yc = 0
    for i in range(1,Nsteps+1):
        xc+=(random.random()-0.5)*2
        yc+=(random.random()-0.5)*2
        x.append(xc)
        y.append(yc)
    return x,y

Rave = []
#function for finding the average length of 1 random walk.
def random_ave(xcor,ycor):
    Nf = len(xcor)-1
    #print(Nf)
    dxi = xcor[0]
    dyi = ycor[0]
    dxf = xcor[Nf]
    dyf = ycor[Nf]
    Rave = np.sqrt((dyf-dyi)**2+(dxf-dxi)**2)
    #print(Rave)
    
    return Rave

rave = []
aveRave = 0
#finds the average length of each walk in 100 walks, then finds the average of that.
for j in range(100):
    xcor,ycor = random_walk(100)
    rave = random_ave(xcor,ycor)
    aveRave += (rave)/100
    rootN = np.sqrt(100)
    plt.plot(xcor,ycor)
plt.show()
print('average displacement of all walks=',aveRave)
print('root of number of steps per walk=',rootN)
#shows the average length of all walks, and the square root of N, N being 100 in this case.
for j in range(100):
    xcor,ycor = random_walk(1000)
    rave = random_ave(xcor,ycor)
    aveRave += (rave)/100
    rootN = np.sqrt(1000)
    plt.plot(xcor,ycor)
plt.show()
print('average displacement of all walks=',aveRave)
print('root of number of steps per walk=',rootN)
#shows the average length of all walks, and the square root of N, N being 1000 in this case.

for j in range(100):
    xcor,ycor = random_walk(10000)
    rave = random_ave(xcor,ycor)
    aveRave += (rave)/100
    rootN = np.sqrt(10000)
    plt.plot(xcor,ycor)
plt.show()
print('average displacement of all walks=',aveRave)
print('root of number of steps per walk=',rootN)
#shows the average length of all walks, and the square root of N, N being 10000 in this case.
```
