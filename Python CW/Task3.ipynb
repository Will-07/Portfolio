```python
import matplotlib.pyplot as plt
import numpy as np
import random

RandNumbers = np.zeros(1000)
RandAve = 0
xtot = 0
for i in range(1000):
    RandNumbers[i] = random.random()
    RandAve += RandNumbers[i]/1000
x = np.zeros(1000)
y = np.zeros(1000)
for i in range(10):
    for j in range(1000):
        x[j] = random.random()
        y[j] = random.random()
    plt.scatter(x,y,s=0.1)
plt.show()
#plot to show distribution of numbers, with random x coord and random y coord, no pattern can be seen.
plt.hist(RandNumbers,bins = 20)
#plot to show distribution of numbers generated in a certain number of bins

def k_moment(k,max,rnd):
    xtot = 0
    res = 1/(k+1)
    for j in range(max):
        xtot += ((1/(max))*rnd[j]**k)
    return xtot,res
#function for finding the kth moment of my set of random numbers

firstmoment = k_moment(1,1000,RandNumbers)
print(firstmoment)
print(RandAve)
#showing that the first moment is approx. equal to the mean

secondmoment = k_moment(2,1000,RandNumbers)
print(secondmoment)
#showing that the variance is high for this set, showing randomness

thirdmoment = k_moment(3,1000,RandNumbers)
print(thirdmoment)
#showing the skew is close to 0.25, sometimes above, sometimes below, meaning the data is symmetric, skewed neither to the left
#or the right.
def nn_check(k,rnd,max):
    ntot = 0
    for i in range(max-k):
        ntot +=(1/(max))*rnd[i]*rnd[i+k]
    return ntot

nn1=nn_check(2,RandNumbers,1000)
print(nn1)
```
