```python
import matplotlib.pyplot as plt
import numpy as np
import random
#same random walk function as task 4
def random_walk1(Nsteps):
    x = []
    y = []
    x.append(0)
    y.append(0)
    xc = 0
    yc = 0
    for i in range(1,Nsteps+1):
        xc+=(random.random()-0.5)*2
        yc+=(random.random()-0.5)*2
        x.append(xc)
        y.append(yc)
    return x,y
#same random walk function but this time with the starting point set a certain distance away
def random_walk2(Nsteps):
    x = []
    y = []
    x.append(50)
    y.append(0)
    xc = 50
    yc = 0
    for i in range(1,Nsteps+1):
        xc+=(random.random()-0.5)*2
        yc+=(random.random()-0.5)*2
        x.append(xc)
        y.append(yc)
    return x,y

def x_dist(x):
    Nf = len(x)-1
    dxi = x[0]
    dxf = x[Nf]
    x_dist = dxf - dxi
    return x_dist
#Thought i needed to make some sort of function to find the point at which the 2 'clouds' cross, but this was proving to be very
#difficult, so decided to just visually analyse them instead.
for j in range(100):
    x1,y1 = random_walk1(300)
    plt.plot(x1,y1)
    plt.xlabel('X')
    plt.ylabel('Y')
    x2,y2 = random_walk2(300)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.plot(x2,y2)
    
    
plt.show()

def random_walk3(Nsteps):
    x = []
    y = []
    x.append(50)
    y.append(0)
    xc = 50
    yc = 0
    for i in range(1,Nsteps+1):
        xc+=(random.random()-0.5)*2
        yc+=(random.random()-0.5)*4
        x.append(xc)
        y.append(yc)
    return x,y
for j in range(100):
    x1,y1 = random_walk1(300)
    plt.plot(x1,y1)
    plt.xlabel('X')
    plt.ylabel('Y')
    x3,y3 = random_walk3(300)
    plt.plot(x3,y3)
plt.show()
```

```python
#the first plot shows that the 2 gasses expanding isotropically will begin to cross at distance 25, or half the seperation. This
#holds for larger seperation, but the larger the seperation, the more walks needed untill they cross.

#the second plot shows that if one of the gasses expands anisotropically, expanding twice as quickly in the y direction, they
#will still cross at a similar point, as they are both expanding at the same rate in the x direction.
```
